```python
%matplotlib inline
# Khai bao cac thu vien ho tro
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
print("pandas.__version__ = ", pd.__version__)
```

```python
# Tao dataframe
# B1: Dinh nghia cac thuoc tinh tinh luong cho nhan su gom:
# * Cac thuoc tinh khong anh huong luong: ID, họ tên, địa chỉ, dân tộc, so thich khong lien quan cong viec
# * Cac thuoc tinh anh huong luong: 
# 1. Bang cap: Loai bang, xep loai, truong dao tao
# 2. Tuoi
# 3. Gioi tinh
# 4. Muc luong cong ty cu tra
# 5. Mang cong viec
# 6. So nam kinh nghiem
# 7. So truong trong cong viec: C++, .NET,...
# 8. Do hot cua cong viec hien tai: A, B, C, D
# 9. Ngoai ngu: English(ielst), English(toiec), Tieng Nhat, Tieng Trung, Tieng Phap
# 10. Vi tri ung tuyen: Teamlead
# 11. He so trach nhiem
# 12. Giai thuong, thanh tich ca nhan: olympic tin hoc, khoa hoc sang tao,...
# 13. Nguong luong cho cong viec: Vd: 7tr- 20tr
# 14. Xếp loại trình độ (năng lực công việc) : A, B, C, D 
# (thuộc tính này cho sẵn. Nếu tính thì có thể dựa vào lịch sử kết quả công việc đã đảm nhận)
# 15. Ý thức, trách nhiệm với công việc, công ty
# 16. Thời gian đã gắn bó với công ty
# 17. Luong (Bai toan so 2)
# * Note: Các thuộc tính cần động, tích lũy hoặc thay đổi theo thời gian

# * Tạo phân phối chuẩn
mu, sigma = 0, 5 # mean and standard deviation
s = np.random.normal(mu, sigma, 300)

# * old_company
old_company = 15

# B2: Tao cac list
len_list = 1000
len_list *= 10

# 1. ID
ID = []
for i in range(len_list):
    temp = "FSI_" + str(i)
    ID.append(temp)
#print(ID)

# 2. Tên
Name = [random.choice(['Harry', 'Ron', 'Hermione', 'Luna', 'Long', 'Nam', 'Lan', 'Ngoc']) for i in range(len_list)]
#print(Name)

# 3. Địa chỉ
Dia_chi = []
for i in range(len_list):
    temp = random.choice(s)
    if(temp <= 5):
        Dia_chi.append('Tinh le')
    else:
        Dia_chi.append('Ha Noi')

# 4. Dân tộc
list_dantoc = ['Thai', 'Mong', 'Tay', 'Nung', 'Muong', 'Dao', 'Hoa', 'Khome', 'Ede', 'Cham']
Dan_toc = []
for i in range(len_list):
    temp = random.choice(s)
    if(temp <= 5):
        Dan_toc.append('Kinh')
    else:
        temp = random.choice(list_dantoc)
        Dan_toc.append(temp)
               
# 5. Sở thích cá nhân
So_thich_ca_nhan = [random.choice(['Nghe nhac', 'xem phim', 'Cau ca', 'Da bong', 'Cau long', 'Gym', 'Yoga']) for i in range(len_list)]
#print(So_thich_ca_nhan)

# 6. Tuổi
Tuoi = []
for i in range(len_list):
    temp = random.choice(s)
    if(temp<=3):
        temp = random.randrange(18, 30, 1)
        Tuoi.append(temp)
    elif(temp>3 and temp<=8):
        temp = random.randrange(30, 40,1)
        Tuoi.append(temp)
    elif(temp>8 and temp<=10):
        temp = random.randrange(40, 50, 1)
        Tuoi.append(temp)
    else:
        temp = random.randrange(50, 60, 1)
        Tuoi.append(temp)

# 7. Giới tính
Gioi_tinh = [random.choice(['Nam', 'Nu']) for i in range(len_list)]

# 8. Chứng chỉ ngoại ngữ
Ngoai_ngu = []
list_language = ['Tieng anh ielts', 'Tieng Nhat', 'Tieng Trung', 'Tieng Phap', 'Tieng Nga']
for i in range(len_list):
    temp = random.choice(s)
    if(temp <= 0):
        Ngoai_ngu.append('Khong chung chi')
    elif(temp>0 and temp<=5):
        Ngoai_ngu.append('Tieng anh toiec')
    else:
        temp = random.choice(list_language)
        Ngoai_ngu.append(temp)

        
# 9. Ý thức trách nhiệm với công việc công ty
Y_thuc_trach_nhiem = [random.choice(['A', 'B', 'C', 'D']) for i in range(len_list)]


# 11, 12, 13. Thông tin về bằng cấp
list_bang = ['Khong', 'Dai hoc', 'Thac si', 'Tien si']
list_truong = ['HNUE', 'HUST', 'EPU', 'VNU', 'USTH', 'HAUI', 'UTC', 'TLU']
Bang_cap = []
Xep_loai_bang = []
Truong_dao_tao = []
for i in range(len_list):
    temp = random.choice(s)
    if(temp<=0):
        bang = random.choice(['Dai hoc', 'Cao dang']);
        Bang_cap.append(bang)
    elif(temp>8):
        if(Tuoi[i] > 27 and Ngoai_ngu[i] != 'Khong chung chi'):
            bang = random.choice(['Thac si', 'Tien si'])
            Bang_cap.append(bang)
        else:
            bang = 'Thac si'
            Bang_cap.append(bang)
    else:
        bang = random.choice(['Dai hoc', 'Cao dang', 'Khong bang cap'])
        Bang_cap.append(bang)
        
    # Khong co bang cap, hoac la bang TS se khong co xep loai
    if(bang == 'Khong bang cap' or bang == 'Tien si'):
        Xep_loai_bang.append('Khong xep loai')
    else:
        Xep_loai_bang.append(random.choice(['Trung binh', 'kha', 'Gioi', 'Xuat sac']))
        
        
    if(bang == 'Khong bang cap'):
        Truong_dao_tao.append('Khong di hoc')
    else:
        temp = random.choice(list_truong)
        Truong_dao_tao.append(temp)

        
# 14. Mảng công việc
Mang_cong_viec = []
for i in range(len_list):
    temp = random.choice(s)
    if(Bang_cap[i] == 'Tien si'):
        Mang_cong_viec.append('Ky su R&D')
    else:
        if(temp <= 0):
            Mang_cong_viec.append('Lap trinh vien')
        elif(temp>0 and temp<=5):
            viec = random.choice(['Ky su R&D', 'BA', 'Tu van ban hang'])
            Mang_cong_viec.append(viec)
        else:
            if(Gioi_tinh[i] == 'Nam'):
                Mang_cong_viec.append('PM')
            else:
                viec = random.choice(['Ke toan', 'Tester', 'Xu ly data'])
                Mang_cong_viec.append(viec)
            
            

# 15. Số năm kinh nghiệm
Nam_kinh_nghiem = []
for i in range(len_list):
    temp = Tuoi[i] - 18;
    if(temp > 0):
        kn = random.randrange(0, temp, 1)
        #print('type = ', type(temp))
        Nam_kinh_nghiem.append(kn)
    else:
        Nam_kinh_nghiem.append(0)
    #print('Nam_kinh_nghiem = ', Nam_kinh_nghiem[i])

# 16. Sở trường công việc
So_truong = []
for i in range(len_list):
    if(Mang_cong_viec[i] == 'Lap trinh vien'):
        temp = random.choice(['C++', '.NET', 'Java', 'Android', 'Ruby', 'PHP', 'Python', 'ios'])
        So_truong.append(temp)
    elif(Mang_cong_viec[i] == 'Ky su R&D'):
        temp = random.choice(['Computer vision', 'NLP', 'Data Science', 'Tools'])
        So_truong.append(temp)
    else:
        So_truong.append('None')

# 10. Mức độ hót của công việc hiện tại
Hot_work = []
for i in range(len_list):
    if(So_truong[i] == 'C++' or So_truong[i] == 'Java' or So_truong[i] == 'ios'):
        Hot_work.append('A')
    elif(So_truong[i] == 'PHP'):
        Hot_work.append('C')
    elif(So_truong[i]=='None'):
        Hot_work.append('D')
    else:
        Hot_work.append('B')

# 17. Vị trí ứng tuyển
Vi_tri = []
for i in range(len_list):
    temp = random.choice(s)
    if(temp < 5):
        Vi_tri.append('Nhan vien')
    else:
        if(Nam_kinh_nghiem[i]>8):
            Vi_tri.append(random.choice(['Technical leader', 'Truong phong']))
        else:
            Vi_tri.append('Teamlead')
    

# 18. Hệ số trách nhiệm
He_so_trach_nhiem = []
for i in range(len_list):
    if(Vi_tri[i] == 'Nhan vien'):
        He_so_trach_nhiem.append(1)
    elif(Vi_tri[i] == 'Teamlead'):
        He_so_trach_nhiem.append(1.5)
    elif(Vi_tri[i] == 'Teachnical leader'):
        He_so_trach_nhiem.append(2)
    elif(Vi_tri[i] == 'Truong phong'):
        He_so_trach_nhiem.append(4)
    else:
        He_so_trach_nhiem.append(1)
        
# 19. Giải thưởng, thành tích cá nhân
Thanh_tich = []
list_thanhtich = ['Oympic toan', 'olympic tin', 'Khoa hoc sang tao', 'Nhan tai dat viet', 'None']
for i in range(len_list):
    temp = random.choice(s)
    if(temp <= 5):
        Thanh_tich.append('None')
    else:
        
        if(Bang_cap[i] == 'Khong'):
            Thanh_tich.append('None')
        else:
            temp = random.choice(list_thanhtich)
            Thanh_tich.append(temp)
# 20. Ngưỡng lương
Nguong_luong = []
for i in range(len_list):
    if(Mang_cong_viec[i] == 'PM'):
        Nguong_luong.append(30)
    elif(Mang_cong_viec[i] == 'Lap trinh vien' or Mang_cong_viec[i] == 'Ky su R&D' or Mang_cong_viec[i] == 'BA'):
        Nguong_luong.append(25)
    else:
        Nguong_luong.append(15)
        
# 21. Xep loai trinh do (Nang luc thuc hien cong viec)
Nang_luc_work = []
for i in range(len_list):
    if(Bang_cap[i]=='Thac si' or Bang_cap[i] == 'Tien si' or Vi_tri[i]=='Truong phong' or Vi_tri[i]== 'Technical leader'):
        temp = random.choice(s)
        if(temp < 5):
            Nang_luc_work.append(random.choice(['A']))
        else:
            Nang_luc_work.append(random.choice(['A', 'B']))
    else:
        Nang_luc_work.append(random.choice(['A', 'B', 'C', 'D']))

# 22. Thời gian gắn bó với công ty
Thoi_gian_gan_bo = []
for i in range(len_list):
    temp = random.choice(s)
    time_max = Tuoi[i] - 18
    if(Tuoi[i] <=19):
        if(temp>=0):
            Thoi_gian_gan_bo.append(0)
        else:
            Thoi_gian_gan_bo.append(1)
    else:
        if(time_max >= old_company):
            Thoi_gian_gan_bo.append(random.randrange(0, old_company, 1))
        else:
            Thoi_gian_gan_bo.append(random.randrange(0, time_max, 1))
        
# 23. Lịch sử lương
Luong_cu = []
for i in range(len_list):
    if(Bang_cap[i] == 'Tien si'):
        money_min = 20
    elif(Bang_cap[i] == 'Thac si'):
        money_min = 15
    elif(Bang_cap[i] == 'Dai hoc '):
        money_min = 7
    else:
        money_min = 4
        
    if(money_min < Nguong_luong[i]):
        temp = random.randrange(money_min, Nguong_luong[i], 1) * He_so_trach_nhiem[i]
        Luong_cu.append(temp)
    else:
        temp = money_min * He_so_trach_nhiem[i]
        Luong_cu.append(temp)
    

# 24. Lương
# Lương mô phỏng phụ thuộc 5 features: Mảng công việc, bằng cấp, số năm kinh nghiệm, vị trí ứng tuyển, năng lực công việc.
Luong = []
for i in range(len_list):
    # 24.1 Ban đầu ước lượng theo mảng công việc
    if(Mang_cong_viec[i] == 'Ke toan' or Mang_cong_viec[i] == 'Xu ly data'):
        base = 4.2
    else:
        base = 7
    
    # 24.2 Xét theo bằng cấp: nếu bằng thạc sĩ ko thấp hơn 10tr, tiến sĩ ko thấp hơn 15tr, các loại khác để nguyên
    if(Bang_cap[i] == 'Thac si'):
        base = 10
    if(Bang_cap[i] == 'Tien si'):
        base = 15
    
    # 24.3 Xét tiếp theo năm kinh nghiệm: 
    # dưới 1 năm kinh nghiệm thì giữ nguyên mức cơ bản
    # 1-3 năm tăng 30% cơ bản
    # 3-5 năm tăng 50% bản
    # Trên 5 năm kinh nghiệm tăng 100% lương
    if(Nam_kinh_nghiem[i]<= 1):
        luong = base
    elif(Nam_kinh_nghiem[i] > 1 and Nam_kinh_nghiem[i]<=3):
        luong = base + base * 0.3
    elif(Nam_kinh_nghiem[i] > 3 and Nam_kinh_nghiem[i]<=5):
        luong = base + base * 0.5
    elif(Nam_kinh_nghiem[i] > 5):
        luong = base * 2
        
    # 24.4 Xét tiếp theo vị trí:
    # Nhân viên: giữ nguyên
    # Leader tăng 150%
    # Teachnical tăng 180%
    # Trưởng phòng tăng 300%
    if(Vi_tri[i] == 'Nhan vien'):
        luong = luong
    elif(Vi_tri[i] == 'Team leader'):
        luong = luong + luong * 0.5
    elif(Vi_tri[i] == 'Technical leader'):
        luong = luong + luong * 0.8
    elif(Vi_tri[i] == 'Truong phong'):
        luong = luong * 3
    
    # 24.5 Xét tiếp theo năng lực
    # Năng lực mức C là bình thường, giữ nguyên
    # Năng lực mức B thêm 2 triệu
    # Năng lực mức A thêm 5 triệu
    # Năng lực mức D kém hơn bình thường, giảm 2 triệu
    if(Nang_luc_work[i] == "C"):
        luong = luong
    elif(Nang_luc_work[i] == "B"):
        luong = luong + 2
    elif(Nang_luc_work[i] == "A"):
        luong = luong + 5
    elif(Nang_luc_work[i] == "D"):
        luong = luong - 2
        
    Luong.append(luong)
        
# B3: Tao dataframe
calculary_salary = pd.DataFrame()
calculary_salary['ID'] = ID
calculary_salary['Name'] = Name
calculary_salary['Dia_chi'] = Dia_chi
calculary_salary['Dan_toc'] = Dan_toc
calculary_salary['So_thich_ca_nhan'] = So_thich_ca_nhan
calculary_salary['Tuoi'] = Tuoi
calculary_salary['Gioi_tinh'] = Gioi_tinh
calculary_salary['Ngoai_ngu'] = Ngoai_ngu
calculary_salary['Y_thuc_trach_nhiem'] = Y_thuc_trach_nhiem
calculary_salary['Hot_work'] = Hot_work

calculary_salary['Bang_cap'] = Bang_cap
calculary_salary['Xep_loai_bang'] = Xep_loai_bang
calculary_salary['Truong_dao_tao'] = Truong_dao_tao

calculary_salary['Mang_cong_viec'] = Mang_cong_viec
calculary_salary['Nam_kinh_nghiem'] = Nam_kinh_nghiem
calculary_salary['So_truong'] = So_truong
calculary_salary['Vi_tri'] = Vi_tri
calculary_salary['He_so_trach_nhiem'] = He_so_trach_nhiem
calculary_salary['Thanh_tich'] = Thanh_tich
calculary_salary['Nguong_luong'] = Nguong_luong
calculary_salary['Nang_luc_work'] = Nang_luc_work
calculary_salary['Thoi_gian_gan_bo'] = Thoi_gian_gan_bo
calculary_salary['Luong_cu'] = Luong_cu

calculary_salary['Luong'] = Luong

# B4: Luu DataFrame
calculary_salary.to_csv('calculary_salary.csv')

print("DONE")
```
