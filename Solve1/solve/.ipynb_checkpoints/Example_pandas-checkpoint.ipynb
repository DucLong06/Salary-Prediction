```python
# 4. Tạo mới dataframe
# Cách 1: Tạo mới từ dictionary
peoples = {'name':['Kim Văn Sáng', 'Lê Thị Trâm'], 'age':[31, 24], 'website': ['https://github.com/sangkv', None]}
df = pd.DataFrame(peoples)
print(df)
```

```python
# Cach 2: Tạo mới dataframe từ các python list
txts = ['Chỗ này chơi cũng khá hay', 'vui, nhất định sẽ quay lại', 'Chỗ này hơi nhiều muỗi']
labels = [1, 1, 0]
df = pd.DataFrame()
df['txt'] = txts
df['label'] = labels
print(df)
```

```python
# 5. Một số thao tác khác trên dataframe

# Sắp xếp dataframe
df = pd.DataFrame({'name':['Nam', 'Long', 'Ngoc', 'Lan'], 'age':[20, 20, 19, 18]})
print("Trước khi sắp xếp\n", df)
df = df.sort_values('age', ascending=True)
print("\n\nSau khi sắp xếp\n", df)
```

```python
# Nối 2 dataframe
df1 = pd.DataFrame({'name': ['Long'], 'age': [20], 'gender':['male']})
df2 = pd.DataFrame({'name': ['Nam', 'Ngoc', 'Lan'], 'age': [20, 19, 18]})
df_new = df1.append(df2, sort=True)
print(df_new)
```

```python
# Xáo trộn các bản ghi trong dataframe

# Xáo trộn các hàng trong df - shuffle dataframe rows
df = pd.DataFrame({'name': ['Long', 'Nam', 'Ngoc', 'Lan'], 'age': [20, 20 , 19, 18]})
print("Trước khi xáo trộn\n", df)
df = df.sample(frac=1).reset_index(drop=True)
print("\n\nSau khi xáo trộn\n", df)
```

```python
# 6. Lưu dataframe về file csv
df.to_csv('comments.csv')
```
